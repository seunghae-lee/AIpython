```python
import matplotlib.pyplot as plt
```

```python
# 예시 데이터: 월별 기온
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
temperatures = [5, 7, 12, 18, 22]
```

```python
#주피터 노트북에서는 변수명만 줘도 데이터가 나온다.
temperatures
```

```python
# 꺾은선 그래프
plt.plot(months,temperatures)
plt.title('Monthly Temperature Change') # 그래프 제목
plt.xlabel('Month')                 # x축 이름
plt.ylabel('Temperature (C)')       # y축 이름
plt.show()                          # 그래프 보여주기
```

```python
# 예시 데이터: 키와 몸무게
height = [160, 165, 170, 175, 180]
weight = [55, 60, 68, 72, 78]
```

```python
#scatter 산점도
plt.scatter(height, weight) # x축 데이터, y축 데이터
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
# 예시 데이터: 과일별 판매량
fruits = ['Apple', 'Banana', 'Orange']
sales = [100, 150, 80]
```

```python
# 예시 데이터: 학생들의 시험 점수
scores = [65, 72, 75, 78, 80, 82, 85, 88, 90, 91, 93, 95, 98, 100]
```

```python
import matplotlib.pyplot as plt

year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
jpn = [523, 529, 536, 532, 527]

plt.figure(figsize=(5, 3))
plt.plot(year,kor, 'r:', label = 'korea') # year 데이터를 x축으로, kor 데이터를 y축으로 하여 빨간색 점선으로 선 그래프를 그리고 범례에 'Korea'라고 표시
plt.plot(year, usa, 'b:', label = 'Usa' )
plt.plot(year, jpn, 'g-', label='Japan')
plt.ylim(400, 600)  # y축 범위 설정
plt.legend()
plt.title('Mathematics performanve')# 그래프 제목을 'Mathematics performance'로 설정합니다.
plt.xlabel('year')
plt.ylabel('score')
plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2000, 2005, 2010, 2015, 2020]
single_person = [15.5, 20.0, 23.9, 27.2, 31.7]
baby_born = [640089, 438707, 470171, 438420, 272337]

plt.figure(figsize=(5,3))
 # 1인 가구 비율(single_person)과 출생아 수(baby_born) 사이의 관계를 나타내는 산점도를 그립니다.
for i in range(len(year)):
  plt.annotate(year[i],xy=(single_person[i], baby_born[i]),xytext=(5,-10),textcoords='offset points') #각 연도에 해당하는 데이터 지점 옆에 해당 연도를 표시하는 주석을 추가
plt.xlim(14,35)
plt.ylim(230000,700000)
plt.xlabel('single-person household ratio')
plt.ylabel('the number of babies born')
plt.show()
```

```python
import matplotlib.pyplot as plt

exercise ={'climbing':0,'soccer':0,'swimming':0,'tennis':0}

print('좋아하는 운동을 입력하세요.')
print('복수 개의 운동을 입력하려면 콤마로 구분하세요.')

while True:
  s = input('climbing,soccer,swimming,tennis,종료: ')
  if s=='종료':
    break
  for a in s.split(','):
    if a in exercise:
      # 해당 운동의 카운트를 1 증가시키는 연산
    else:
      print(f'{a}는 등록되지 않은 운동입니다.')

plt.figure(figsize=(5.5,3))
plt.bar(exercise.keys(), exercise.values(), color='g', width=0.6)
#plt.xticks(rotation='vertical')
plt.xlabel('exercise')
plt.ylabel('frequency')
plt.show()
```

```python
# 연습문제 : 다음 데이터를 이용하여 1년 동안의 월별 평균 강수량을 꺾은선 그래프로 그려보세요.
# 예시 데이터: 월별 기온
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 7, 50, 35]

plt.figure(figsize=(16, 10))
for i in range(len(rainfall)):
  plt.annotate(rainfall[i], xy=(months[i],rainfall[i]), xytext=(5,-10), textcoords='offset points')
plt.ylim(0,350)
plt.plot(months, rainfall,'go-.') #color='g', marker='o', linestyle='-.'
plt.show()
```
